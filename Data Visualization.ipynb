## Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Line plot
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label="Apple") #to change color of the line use color parameter in plot
plt.style.use("tableau-colorblind10") # choose any available themes
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x,y2,color='green',label="Kiwi",marker="*")
plt.legend() # it will show name of lines at to left corner

plt.show() #it also shows that you have finished one plot

```

```python
## to check the available themes
themes=plt.style.available
print(themes)
```

```python
plt.plot(x,y1,'bo',color='red') #to change color of the line use color parameter in plot
plt.style.use("tableau-colorblind10") # choose any available themes
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()   #since we are using plt.show( two times so two graphs should be produced)
plt.plot(x,y2,color='green',linestyle="dashed")
plt.xlabel("Time",color='green')
plt.ylabel("work",color='green')
plt.title("Time-Work Graph",color="green")


plt.show() #it also shows that you have finished one plot

```

## Scatter Plot

```python
# adjust the size of the plot
plt.figure(figsize=(5,10))


plt.scatter(x,y1,color='red',label="Apple",marker="^") #to change color of the line use color parameter in plot
plt.style.use("tableau-colorblind10") # choose any available themes
plt.xlabel("x values")
plt.ylabel("y values")
plt.scatter(x,y2,color='green',label="Kiwi",marker="*")
plt.legend() # it will show name of lines at to left corner
plt.title("Scatter plot")
plt.show() #it also shows that you have finished one plot

```

## Bar graphs

```python
plt.bar([0,1,2],[10,20,30]) ##current year
plt.bar([0,1,2],[20,10,30]) ##another year , use to compare te graphs

plt.show()
```

```python
# to compare two graphs side by side, change the x coordinate with some value
x_coo=np.array([0,1,2])*2
plt.bar(x_coo,[10,20,15],width=0.5,label="Current Year",tick_label=['gold','silver','platinum'])
plt.bar(x_coo+0.5,[13,43,15],width=0.5,label="Next Year")
plt.legend()
plt.title("Metal Price Comaprison")
plt.xlabel("Metals")
plt.ylabel("Value in price")
plt.show()
#tick_label gives the name to thebars and value is passed in list
```

```python
# to compare two graphs side by side, change the x coordinate with some value
plt.style.use("dark_background")
x_coo=np.array([0,1,2])*2
plt.bar(x_coo,[10,20,15],width=0.5,label="Current Year",tick_label=['gold','silver','platinum'],color='orange')
plt.bar(x_coo+0.5,[13,43,15],width=0.5,label="Next Year",color='green')
plt.legend()
## to give limit to y axis
plt.ylim(0,30)
plt.xlim(0,10)
plt.title("Metal Price Comaprison")
plt.xlabel("Metals")
plt.ylabel("Value in price")
plt.show()
#tick_label gives the name to thebars and value is passed in list
# x_coo is x cordinates and next to it is height of the bar
```
